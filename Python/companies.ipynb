```python
import pandas as pd
import matplotlib.pyplot as plt
path = r'C:\Users\Boyan1\Documents\GitHub\LUISS\Python\Largest-Companies - 25apr(2024)TTM.csv'

pd.set_option('display.max_columns', None)
df = pd.read_csv(path, encoding='ISO-8859-1')


df["asset_turnover"] = df["Sales"] / df["Assets"]


for col in df.columns: 
    if df[col].dtype in ['int64', 'float64'] and (col == "Sales" or col == "Assets" or col == "Market Value" or col == "asset_turnover"):
        print(f"{col} max: {round(df[col].max(), 4)}")
        print(f"{col} min: {round(df[col].min(), 4)}")
        print(f"{col} mean: {round(df[col].mean(), 4)}")
        plt.figure(figsize=(6,4))
        plt.hist(df[col].dropna(), bins=30, edgecolor="black")
        plt.title(f"Histogram of {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.show()

        









    
   
    
```
